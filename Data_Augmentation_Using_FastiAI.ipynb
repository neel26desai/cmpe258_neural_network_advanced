```python
from fastai.vision.all import *
```

```python
TEST_IMAGE='./download.jpeg'

im = PILImage(PILImage.create(TEST_IMAGE))
```

```python
im
```

```python
#Using RandomTransformer
RandTransform(split_idx = 0)(im,split_idx=0)
```

```python
#Flipping the Image
FlipItem(p=1.)(im,split_idx=0)
```

```python
#flippinf the same image to  multiple extend
_,axs = subplots(2, 4)
for ax in axs.flatten():
    show_image(DihedralItem(p=1.)(im, split_idx=0), ctx=ax)
```

```python
#resizing the image
_,axs = plt.subplots(1,3,figsize=(12,4))
for ax,sz in zip(axs.flatten(), [300, 500, 700]):
    show_image(im.crop_pad(sz), ctx=ax, title=f'Size {sz}');
    print(im.crop_pad(sz).shape)
```

```python
#trying different padding options
_,axs = plt.subplots(1,3,figsize=(12,4))
for ax,mode in zip(axs.flatten(), [PadMode.Zeros, PadMode.Border, PadMode.Reflection]):
    show_image(im.crop_pad((600,700), pad_mode=mode), ctx=ax, title=mode);
```

```python
#random cropping
_,axs = plt.subplots(1,3,figsize=(12,4))
f = RandomCrop(200)
for ax in axs: show_image(f(im), ctx=ax);
```
