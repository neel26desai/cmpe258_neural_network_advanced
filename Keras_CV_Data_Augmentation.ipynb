```python
import os
os.environ["KERAS_BACKEND"] = "tensorflow"  # Or "jax" or "torch"!

import tensorflow as tf
import keras_cv
import tensorflow_datasets as tfds
import keras
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
```

```python
import numpy as np
```

```python
# Load your image
image_path = '/content/lotus-flower-doodle-icon-illustration-on-white-simple-drawing-in-blue-color-vector.jpg'  # Replace with your image path
image = load_img(image_path)  # Load the image
image = img_to_array(image)  # Convert the image to a numpy array
image = np.expand_dims(image, axis=0)  # Add batch dimension
```

```python
np.max(image)
```

```python
plt.imshow(image[0])
```

```python
# prompt: apply different augmentationns tothe image using keras cv and print each image

import matplotlib.pyplot as plt
# Apply different augmentations to the image using Keras CV
augmented_images = []
titles = []
# Apply rotation augmentation
augmented_image = keras_cv.layers.RandomRotation(0.5)(image)
augmented_images.append(augmented_image)
titles.append('Rotation')


# Apply some randomAugmentation
augmented_image = keras_cv.layers.RandAugment(value_range=(0, 255))(image)
augmented_images.append(augmented_image)
titles.append('Random Augmentation')

# Apply contrast augmentation
augmented_image = keras_cv.layers.AutoContrast(value_range=(0,255))(image)
augmented_images.append(augmented_image)
titles.append('Auto Contrast')

# Apply zoom augmentation
augmented_image = keras_cv.layers.RandomZoom(0.5)(image)
augmented_images.append(augmented_image)
titles.append('Random Zoom')

# Apply shear augmentation
#Apply shear Augmentation
augmented_image = keras_cv.layers.RandomShear(0.5)(image)
augmented_images.append(augmented_image)
titles.append('Random Shear')
```

```python

import matplotlib.pyplot as plt
for i in range(len(augmented_images)):
  plt.figure()
  plt.imshow(augmented_images[i][0].numpy())
  plt.title(titles[i])
  plt.show()
```
